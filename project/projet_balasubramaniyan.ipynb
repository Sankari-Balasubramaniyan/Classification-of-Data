# MODEL DESIGN

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression, SGDClassifier, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
sgd = SGDClassifier()
lr = LogisticRegression()
dt = DecisionTreeClassifier(max_depth=20)
knn = KNeighborsClassifier(5)
methods = [gnb, dt, sgd, knn ,lr]
```

```python
tps_fit = {}
tps_pred = {}
for m in methods:
    tps_fit[m] = []
    tps_pred[m] = []
for m in methods:
    tf = %timeit -o -q -r1 -n1 m.fit(X_train, y_train)
    tp = %timeit -o -q -r1 -n1 m.predict(X_train)
    tps_fit[m].append(tf.timings[0])
    tps_pred[m].append(tp.timings[0])
    print("{:>25} fit : {:1.3f} ; Pred {:1.3f}".format(m.__class__.__name__,tf.timings[0],tp.timings[0]))
    print("-"*70) 
```

### PREDICTION AND SCORE

```python
#Predictions and score for training dataset

predictions_train = {}
scores_train = {}
for m in methods:
    predictions_train[m] = (m.predict(X_train))
    scores_train[m] = (m.score(X_train,y_train))
```

```python
for m in methods:
    print("{:>20}: {:.3f}".format(m.__class__.__name__,m.score(X_train, y_train)))
```

```python
#Predictions and test for test dataset

predictions_test = {}
scores_test = {}
for m in methods:
    predictions_test[m] = (m.predict(X_test))
    scores_test[m] = (m.score(X_test,y_test))
```

```python
for m in methods:
    print("{:>20}: {:.3f}".format(m.__class__.__name__,m.score(X_test, y_test)))
```

# PERFORMANCE EVALUATION

```python
from sklearn import metrics

ACC = {}           #Accuracy
r2 = {}            #Explained variance
MAE = {}           #Mean Absolute Error
RMSE = {}          #Root Mean Squared Error
MSLE = {}          #Mean squared log error
Ac = []

for m in methods:
    ACC[m] = metrics.accuracy_score(y_test, predictions_test[m])
    Ac.append(metrics.accuracy_score(y_test, predictions_test[m]))
    r2[m] = metrics.r2_score(y_test, predictions_test[m])
    MAE[m] = metrics.mean_absolute_error(y_test, predictions_test[m])
    RMSE[m] = np.sqrt(metrics.mean_squared_error(y_test, predictions_test[m]))
    MSLE[m] = metrics.mean_squared_log_error(y_test, predictions_test[m])
    
```

```python
for m in methods:
    print(m)
    print("Accuracy:",round(ACC[m], 2))
    print("R2 (explained variance):", round(r2[m], 2))
    print("Mean Absolute Error (Σ|y-pred|/n):", round(MAE[m],2))
    print("Root Mean Squared Error (sqrt(Σ(y-pred)^2/n)):", round(RMSE[m],2))
    print("Mean squared log error:", round(MSLE[m], 2))
    print("-"*70)
          
          
```

# HYPERPARAMETER

```python
#Tunning hyperparameter for Decision Tree classifier

from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition

std_slc = StandardScaler()
pca = decomposition.PCA()

pipe = Pipeline(steps=[('std_slc', std_slc),('dec_tree', dt)])
n_components = list(range(1,X_train.shape[1]+1,1))
criterion = ['gini', 'entropy']
max_depth = [2,4,6,8,10,12]
parameters = dict(dec_tree__criterion=criterion,dec_tree__max_depth=max_depth)
clf_GS = GridSearchCV(pipe, parameters)
clf_GS.fit(X_train, y_train)
print('Best Criterion:', clf_GS.best_estimator_.get_params()['dec_tree__criterion'])
print('Best max_depth:', clf_GS.best_estimator_.get_params()['dec_tree__max_depth'])
print(clf_GS.best_estimator_.get_params()['dec_tree'])    
```

```python
#Learning curve for GaussianNB

from sklearn.model_selection import learning_curve
    
ylim=(0.5, 1.01)
cv=None
train_sizes=np.linspace(.1, 1.0, 5)
n_jobs=4
train_sizes, train_scores, test_scores = learning_curve(GaussianNB(), X_train, y_train, 
                                                         cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)      
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
```

```python
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")

plt.xlabel("Training examples")
plt.ylabel("Score")

plt.show()
```
